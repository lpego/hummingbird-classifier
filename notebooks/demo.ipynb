```python
import os
import urllib.request
import zipfile
import shutil

# Define URLs
demo_data_url = "https://gitlab.renkulab.io/biodetect/hummingbird-classifier-models/-/archive/main/hummingbird-classifier-models-main.zip?path=data/demo"
models_url = "https://gitlab.renkulab.io/biodetect/hummingbird-classifier-models/-/archive/main/hummingbird-classifier-models-main.zip?path=models/mobilenet-v0"

# Define paths
workdir = "/home/jovyan/work/"
demo_data_zip_path = os.path.join(workdir, "hummingbird-classifier/data/demo/demo_data.zip")
models_zip_path = os.path.join(workdir, "hummingbird-classifier/models/mobilenet-v0/models.zip")
demo_data_extract_path = os.path.join(workdir, "hummingbird-classifier/data/demo")
models_extract_path = os.path.join(workdir, "hummingbird-classifier/models/mobilenet-v0")

# Create directories if they don't exist
os.makedirs(demo_data_extract_path, exist_ok=True)
os.makedirs(models_extract_path, exist_ok=True)

# Download demo data
urllib.request.urlretrieve(demo_data_url, demo_data_zip_path)

# Download models
urllib.request.urlretrieve(models_url, models_zip_path)

# Unzip demo data
with zipfile.ZipFile(demo_data_zip_path, 'r') as zip_ref:
    zip_ref.extractall(demo_data_extract_path)

# Move demo data to correct location
extracted_demo_data_path = os.path.join(demo_data_extract_path, 'hummingbird-classifier-models-main-data-demo/data/demo')
for item in os.listdir(extracted_demo_data_path):
    s = os.path.join(extracted_demo_data_path, item)
    d = os.path.join(demo_data_extract_path, item)
    if os.path.isdir(s):
        if not os.path.exists(d):
            os.rename(s, d)
    else:
        os.replace(s, d)

# Remove old demo data folder and archive
shutil.rmtree(os.path.join(demo_data_extract_path, 'hummingbird-classifier-models-main-data-demo'))
os.remove(demo_data_zip_path)

# Unzip models
with zipfile.ZipFile(models_zip_path, 'r') as zip_ref:
    zip_ref.extractall(models_extract_path)

# Move models to correct location
extracted_models_path = os.path.join(models_extract_path, 'hummingbird-classifier-models-main-models-mobilenet-v0/models')
for item in os.listdir(extracted_models_path):
    s = os.path.join(extracted_models_path, item)
    d = os.path.join(models_extract_path, item)
    if os.path.isdir(s):
        if not os.path.exists(d):
            os.rename(s, d)
    else:
        os.replace(s, d)
        
# Remove old models folder and archive
shutil.rmtree(os.path.join(models_extract_path, 'hummingbird-classifier-models-main-models-mobilenet-v0'))
os.remove(models_zip_path)
```
