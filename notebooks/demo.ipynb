```python
import os
import urllib.request
import zipfile
import shutil

# Define URLs
demo_data_url = "https://gitlab.renkulab.io/biodetect/hummingbird-classifier-models/-/archive/main/hummingbird-classifier-models-main.zip?path=data/demo"
models_url = "https://gitlab.renkulab.io/biodetect/hummingbird-classifier-models/-/archive/main/hummingbird-classifier-models-main.zip?path=models/mobilenet-v0"

# Define paths
workdir = ".."
temp = os.path.join(workdir, 'temp')
demo_data_zip_path = os.path.join(workdir, "temp/demo_data.zip")
models_zip_path = os.path.join(workdir, "temp/models.zip")
# demo_data_destination = os.path.join(workdir, "data/demo")
# demo_models_destination = os.path.join(workdir, "models/mobilenet-v0")

# # Create directories if they don't exist
# os.makedirs(demo_data_extract_path, exist_ok=True)
# os.makedirs(models_extract_path, exist_ok=True)
os.makedirs(temp, exist_ok=True)

# Download demo data
urllib.request.urlretrieve(demo_data_url, demo_data_zip_path)

# Download models
urllib.request.urlretrieve(models_url, models_zip_path)

# Unzip demo data
with zipfile.ZipFile(demo_data_zip_path, 'r') as zip_ref:
    zip_ref.extractall(temp)

# Unzip models
with zipfile.ZipFile(models_zip_path, 'r') as zip_ref:
    zip_ref.extractall(temp)

# Move demo data to correct location
# extracted_demo_data_path = os.path.join(workdir, 'temp', 'hummingbird-classifier-models-main-data-demo')
# os.makedirs(extracted_demo_data_path, exist_ok=True)
# for item in os.listdir(extracted_demo_data_path):
#     s = os.path.join(extracted_demo_data_path, item)
#     d = os.path.join(demo_data_extract_path, item)
#     if os.path.isdir(s):
#         if not os.path.exists(d):
#             os.rename(s, d)
#     else:
#         os.replace(s, d)
shutil.copytree(os.path.join(temp, 'hummingbird-classifier-models-main-data-demo/data/demo'), os.path.join(workdir, 'data/demo'), dirs_exist_ok=True)
    
# Move models to correct location
# extracted_models_path = os.path.join(workdir, 'temp', 'hummingbird-classifier-models-main-models-mobilenet-v0')
# os.makedirs(extracted_models_path, exist_ok=True)
# for item in os.listdir(extracted_models_path):
#     s = os.path.join(extracted_models_path, item)
#     d = os.path.join(models_extract_path, item)
#     if os.path.isdir(s):
#         if not os.path.exists(d):
#             os.rename(s, d)
#     else:
#         os.replace(s, d)
shutil.copytree(os.path.join(temp, 'hummingbird-classifier-models-main-models-mobilenet-v0/models/mobilenet-v0'), os.path.join(workdir, 'models/mobilenet-v0'), dirs_exist_ok=True)

# Remove old archives and extracted files
# shutil.rmtree(temp)
```
